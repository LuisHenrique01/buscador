```python
import requests as re
from bs4 import BeautifulSoup as bs
import pandas as pd
url_inicial = "https://pt.wikipedia.org/wiki/Wikip%C3%A9dia:P%C3%A1gina_principal"
```

```python
def retorna_links(url=url_inicial):
    response = re.get(url)
    soup = bs(response.text, 'html5lib')
    links = soup.find_all('a')
    #so_links = []
    #for link in links:
    #    try:
    #        if '/wiki/' in link['href']: 
    #            so_links.append(link['href'])
    #    except KeyError:
    #        pass
    return links
```

```python
def busca(chave, profundidade, links):
    resultados = [[links, 'Home']]
    links = [links]
    for _ in range(profundidade):
        if profundidade != 0:
            for link, titulo in resultados:
                links = retorna_links(link)
            for link in links:
                try:
                    if link['href'][:6] == '/wiki/':
                        resultados.append(["%s"%('https://pt.wikipedia.org' + link['href']), link['title']])
                    elif '/wiki/' in link['href'] and link['href'][:10] == 'htpps://pt':
                        resultados.append([link['href'], link['title']])
                except KeyError:
                    pass
    return resultados
```
